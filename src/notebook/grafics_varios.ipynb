```python
# Obtener los cuartiles y la mediana para el coste de reparación por tipo de defecto y severidad
q1 = df.groupby(['defect_type', 'severity'])['repair_cost'].quantile(0.25)
q3 = df.groupby(['defect_type', 'severity'])['repair_cost'].quantile(0.75)
median = df.groupby(['defect_type', 'severity'])['repair_cost'].median()

# Mostrar los resultados
print(f"Primer cuartil:\n{q1}")
print(f"Tercer cuartil:\n{q3}")
print(f"Mediana:\n{median}")
```

```python

# Crear el gráfico de cajas
plt.figure(figsize=(10, 6))
ax = sns.boxplot(x='defect_type', y='repair_cost', hue='severity', data=df, palette='Set2')

# Extraer las estadísticas del gráfico
for i, box in enumerate(ax.artists):
    # Obtener los valores de los cuartiles
    median = ax.lines[2*i+1].get_ydata()[0]
    lower = ax.lines[2*i].get_ydata()[0]
    upper = ax.lines[2*i+2].get_ydata()[0]
    print(f'Box {i+1}:')
    print(f'  Mediana: {median}')
    print(f'  Cuartil inferior: {lower}')
    print(f'  Cuartil superior: {upper}')

plt.tight_layout()
plt.show()
```

```python
# Calcular la desviación estándar y el coeficiente de variación por tipo de defecto y severidad
desviacion_estandar = df.groupby(['defect_type', 'severity'])['repair_cost'].std()
media = df.groupby(['defect_type', 'severity'])['repair_cost'].mean()
coef_variacion = desviacion_estandar / media

# Crear un DataFrame con estos resultados
estadisticas = pd.DataFrame({
    'Desviación Estándar': desviacion_estandar,
    'Coef. de Variación': coef_variacion
})

# Mostrar las estadísticas
estadisticas
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Estadísticas descriptivas del coste de reparación agrupado por 'defect_type'
repair_cost_by_defect_type = df.groupby('defect_type')['repair_cost'].describe()

# Estadísticas descriptivas del coste de reparación agrupado por 'severity'
repair_cost_by_severity = df.groupby('severity')['repair_cost'].describe()

# Estadísticas descriptivas del coste de reparación agrupado por 'defect_location'
repair_cost_by_defect_location = df.groupby('defect_location')['repair_cost'].describe()

# Crear una figura
fig, ax = plt.subplots(figsize=(10, 6))

# Crear una tabla visual para cada grupo (usando .values para obtener los valores de la estadística)
ax.axis('tight')
ax.axis('off')

# Concatenamos las tres tablas en una sola para mostrar
df_combined = pd.concat([repair_cost_by_defect_type, repair_cost_by_severity, repair_cost_by_defect_location], keys=["Defect Type", "Severity", "Defect Location"])

# Mostrar la tabla en el gráfico
table = ax.table(cellText=df_combined.values, colLabels=df_combined.columns, loc='center')

plt.show()
```

```python
import seaborn as sns

# Estadísticas descriptivas por tipo de defecto (ejemplo de "mean" y "std")
repair_cost_by_defect_type = df.groupby('defect_type')['repair_cost'].agg(['mean', 'std']).reset_index()

# Crear gráfico de barras
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Gráfico para 'mean' por 'defect_type'
sns.barplot(x='defect_type', y='mean', data=repair_cost_by_defect_type, ax=axes[0], palette='Blues_d')
axes[0].set_title('Promedio de Coste de Reparación por Tipo de Defecto')
axes[0].set_ylabel('Promedio')

# Gráfico para 'std' por 'defect_type'
sns.barplot(x='defect_type', y='std', data=repair_cost_by_defect_type, ax=axes[1], palette='Greens_d')
axes[1].set_title('Desviación Estándar del Coste de Reparación por Tipo de Defecto')
axes[1].set_ylabel('Desviación Estándar')

# Gráfico para 'mean' por 'defect_location'
repair_cost_by_defect_location = df.groupby('defect_location')['repair_cost'].agg(['mean', 'std']).reset_index()
sns.barplot(x='defect_location', y='mean', data=repair_cost_by_defect_location, ax=axes[2], palette='Purples_d')
axes[2].set_title('Promedio de Coste de Reparación por Localización de Defecto')
axes[2].set_ylabel('Promedio')

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Estadísticas descriptivas por 'defect_type'
repair_cost_by_defect_type = df.groupby('defect_type')['repair_cost'].agg(['mean', 'std', 'min', 'max']).reset_index()

# Crear gráficos para cada estadística
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Gráfico de la Media
sns.barplot(x='defect_type', y='mean', data=repair_cost_by_defect_type, ax=axes[0, 0], palette='Blues_d')
axes[0, 0].set_title('Media del Coste de Reparación por Tipo de Defecto')
axes[0, 0].set_ylabel('Media')

# Gráfico de la Desviación Estándar
sns.barplot(x='defect_type', y='std', data=repair_cost_by_defect_type, ax=axes[0, 1], palette='Greens_d')
axes[0, 1].set_title('Desviación Estándar del Coste de Reparación por Tipo de Defecto')
axes[0, 1].set_ylabel('Desviación Estándar')

# Gráfico del valor Máximo
sns.barplot(x='defect_type', y='max', data=repair_cost_by_defect_type, ax=axes[1, 0], palette='Reds_d')
axes[1, 0].set_title('Valor Máximo del Coste de Reparación por Tipo de Defecto')
axes[1, 0].set_ylabel('Máximo')

# Gráfico del valor Mínimo
sns.barplot(x='defect_type', y='min', data=repair_cost_by_defect_type, ax=axes[1, 1], palette='Purples_d')
axes[1, 1].set_title('Valor Mínimo del Coste de Reparación por Tipo de Defecto')
axes[1, 1].set_ylabel('Mínimo')

# Ajustar el layout para que no se solapen
plt.tight_layout()

# Mostrar los gráficos
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Estadísticas descriptivas por 'defect_type'
repair_cost_by_defect_type = df.groupby('defect_type')['repair_cost'].agg(['mean', 'std', 'min', 'max', 'quantile']).reset_index()

# Calcular el ICQ (Q3 - Q1)
repair_cost_by_defect_type['ICQ'] = repair_cost_by_defect_type['quantile'].apply(lambda x: x[0.75] - x[0.25])

# Crear gráficos para cada estadística
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Gráfico de la Media
sns.barplot(x='defect_type', y='mean', data=repair_cost_by_defect_type, ax=axes[0, 0], palette='Blues_d')
axes[0, 0].set_title('Media del Coste de Reparación por Tipo de Defecto')
axes[0, 0].set_ylabel('Media')

# Gráfico del ICQ (Interquartile Range)
sns.barplot(x='defect_type', y='ICQ', data=repair_cost_by_defect_type, ax=axes[0, 1], palette='Purples_d')
axes[0, 1].set_title('Rango Intercuartil (ICQ) del Coste de Reparación por Tipo de Defecto')
axes[0, 1].set_ylabel('ICQ')

# Gráfico del valor Máximo
sns.barplot(x='defect_type', y='max', data=repair_cost_by_defect_type, ax=axes[1, 0], palette='Reds_d')
axes[1, 0].set_title('Valor Máximo del Coste de Reparación por Tipo de Defecto')
axes[1, 0].set_ylabel('Máximo')

# Gráfico del valor Mínimo
sns.barplot(x='defect_type', y='min', data=repair_cost_by_defect_type, ax=axes[1, 1], palette='Greens_d')
axes[1, 1].set_title('Valor Mínimo del Coste de Reparación por Tipo de Defecto')
axes[1, 1].set_ylabel('Mínimo')

# Ajustar el layout para que no se solapen
plt.tight_layout()

# Mostrar los gráficos
plt.show()
```
