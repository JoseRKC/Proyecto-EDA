```python
# Obtener los cuartiles y la mediana para el coste de reparacin por tipo de defecto y severidad
q1 = df.groupby(['defect_type', 'severity'])['repair_cost'].quantile(0.25)
q3 = df.groupby(['defect_type', 'severity'])['repair_cost'].quantile(0.75)
median = df.groupby(['defect_type', 'severity'])['repair_cost'].median()

# Mostrar los resultados
print(f"Primer cuartil:\n{q1}")
print(f"Tercer cuartil:\n{q3}")
print(f"Mediana:\n{median}")
```

```python

# Crear el grfico de cajas
plt.figure(figsize=(10, 6))
ax = sns.boxplot(x='defect_type', y='repair_cost', hue='severity', data=df, palette='Set2')

# Extraer las estadsticas del grfico
for i, box in enumerate(ax.artists):
    # Obtener los valores de los cuartiles
    median = ax.lines[2*i+1].get_ydata()[0]
    lower = ax.lines[2*i].get_ydata()[0]
    upper = ax.lines[2*i+2].get_ydata()[0]
    print(f'Box {i+1}:')
    print(f'  Mediana: {median}')
    print(f'  Cuartil inferior: {lower}')
    print(f'  Cuartil superior: {upper}')

plt.tight_layout()
plt.show()
```

```python
# Calcular la desviacin estndar y el coeficiente de variacin por tipo de defecto y severidad
desviacion_estandar = df.groupby(['defect_type', 'severity'])['repair_cost'].std()
media = df.groupby(['defect_type', 'severity'])['repair_cost'].mean()
coef_variacion = desviacion_estandar / media

# Crear un DataFrame con estos resultados
estadisticas = pd.DataFrame({
    'Desviacin Estndar': desviacion_estandar,
    'Coef. de Variacin': coef_variacion
})

# Mostrar las estadsticas
estadisticas
```

```python
import seaborn as sns

# Estadsticas descriptivas por tipo de defecto (ejemplo de "mean" y "std")
repair_cost_by_defect_type = df.groupby('defect_type')['repair_cost'].agg(['mean', 'std']).reset_index()

# Crear grfico de barras
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Grfico para 'mean' por 'defect_type'
sns.barplot(x='defect_type', y='mean', data=repair_cost_by_defect_type, ax=axes[0], palette='Blues_d')
axes[0].set_title('Promedio de Coste de Reparacin por Tipo de Defecto')
axes[0].set_ylabel('Promedio')

# Grfico para 'std' por 'defect_type'
sns.barplot(x='defect_type', y='std', data=repair_cost_by_defect_type, ax=axes[1], palette='Greens_d')
axes[1].set_title('Desviacin Estndar del Coste de Reparacin por Tipo de Defecto')
axes[1].set_ylabel('Desviacin Estndar')

# Grfico para 'mean' por 'defect_location'
repair_cost_by_defect_location = df.groupby('defect_location')['repair_cost'].agg(['mean', 'std']).reset_index()
sns.barplot(x='defect_location', y='mean', data=repair_cost_by_defect_location, ax=axes[2], palette='Purples_d')
axes[2].set_title('Promedio de Coste de Reparacin por Localizacin de Defecto')
axes[2].set_ylabel('Promedio')

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Estadsticas descriptivas por 'defect_type'
repair_cost_by_defect_type = df.groupby('defect_type')['repair_cost'].agg(['mean', 'std', 'min', 'max']).reset_index()

# Crear grficos para cada estadstica
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Grfico de la Media
sns.barplot(x='defect_type', y='mean', data=repair_cost_by_defect_type, ax=axes[0, 0], palette='Blues_d')
axes[0, 0].set_title('Media del Coste de Reparacin por Tipo de Defecto')
axes[0, 0].set_ylabel('Media')

# Grfico de la Desviacin Estndar
sns.barplot(x='defect_type', y='std', data=repair_cost_by_defect_type, ax=axes[0, 1], palette='Greens_d')
axes[0, 1].set_title('Desviacin Estndar del Coste de Reparacin por Tipo de Defecto')
axes[0, 1].set_ylabel('Desviacin Estndar')

# Grfico del valor Mximo
sns.barplot(x='defect_type', y='max', data=repair_cost_by_defect_type, ax=axes[1, 0], palette='Reds_d')
axes[1, 0].set_title('Valor Mximo del Coste de Reparacin por Tipo de Defecto')
axes[1, 0].set_ylabel('Mximo')

# Grfico del valor Mnimo
sns.barplot(x='defect_type', y='min', data=repair_cost_by_defect_type, ax=axes[1, 1], palette='Purples_d')
axes[1, 1].set_title('Valor Mnimo del Coste de Reparacin por Tipo de Defecto')
axes[1, 1].set_ylabel('Mnimo')

# Ajustar el layout para que no se solapen
plt.tight_layout()

# Mostrar los grficos
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Estadsticas descriptivas por 'defect_type'
repair_cost_by_defect_type = df.groupby('defect_type')['repair_cost'].agg(['mean', 'std', 'min', 'max', 'quantile']).reset_index()

# Calcular el ICQ (Q3 - Q1)
repair_cost_by_defect_type['ICQ'] = repair_cost_by_defect_type['quantile'].apply(lambda x: x[0.75] - x[0.25])

# Crear grficos para cada estadstica
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Grfico de la Media
sns.barplot(x='defect_type', y='mean', data=repair_cost_by_defect_type, ax=axes[0, 0], palette='Blues_d')
axes[0, 0].set_title('Media del Coste de Reparacin por Tipo de Defecto')
axes[0, 0].set_ylabel('Media')

# Grfico del ICQ (Interquartile Range)
sns.barplot(x='defect_type', y='ICQ', data=repair_cost_by_defect_type, ax=axes[0, 1], palette='Purples_d')
axes[0, 1].set_title('Rango Intercuartil (ICQ) del Coste de Reparacin por Tipo de Defecto')
axes[0, 1].set_ylabel('ICQ')

# Grfico del valor Mximo
sns.barplot(x='defect_type', y='max', data=repair_cost_by_defect_type, ax=axes[1, 0], palette='Reds_d')
axes[1, 0].set_title('Valor Mximo del Coste de Reparacin por Tipo de Defecto')
axes[1, 0].set_ylabel('Mximo')

# Grfico del valor Mnimo
sns.barplot(x='defect_type', y='min', data=repair_cost_by_defect_type, ax=axes[1, 1], palette='Greens_d')
axes[1, 1].set_title('Valor Mnimo del Coste de Reparacin por Tipo de Defecto')
axes[1, 1].set_ylabel('Mnimo')

# Ajustar el layout para que no se solapen
plt.tight_layout()

# Mostrar los grficos
plt.show()
```

```python
# Usar One-Hot Encoding para columnas categricas, como 'defect_type'
df = pd.get_dummies(df, columns=['defect_type'])
```
